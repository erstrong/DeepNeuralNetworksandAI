```python
import pandas as pd
import numpy as np
import pickle

from tpot import TPOTClassifier
from autosklearn.classification import AutoSklearnClassifier

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix

import warnings
warnings.filterwarnings(action='once')
```

```python
raw_scores = pd.read_csv('data_files/output_for.csv')
```

```python
raw_scores.info()
```

```python
raw_scores.head()
```

```python
score_columns = ['IbmScore', 'GCPScore', 'AzureScore', 'AWSPosScore', 'AWSNegScore', 'AWSNeuScore']

x_data = np.asarray(raw_scores[score_columns])
y_data = np.asarray(raw_scores['ActualLabel'])
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.2, random_state=11)
```

```python
x_train.shape
```

```python
y_train.shape
```

# TPOT

```python
tpot = TPOTClassifier(generations=5, population_size=20, verbosity=2)
tpot.fit(x_train, y_train)
```

```python
predictions = tpot.predict(x_test)
```

```python
predictions
```

```python
accuracy_score(y_test, predictions)
```

```python
confusion_matrix(y_test, predictions)
```

# Auto-Sklearn

```python
y_train_dummies = pd.get_dummies(y_train)
y_test_dummies = pd.get_dummies(y_test)
```

```python
automl = AutoSklearnClassifier()
automl.fit(x_train, np.asarray(y_train_dummies))
```

```python
predictions = automl.predict(x_test)
```

```python
accuracy_score(y_test_dummies, predictions)
```

```python
def convert(x):
    out = []
    for item in x:
        if item == 'positive':
            out.append(2)
        elif item == 'neutral':
            out.append(1)
        else:
            out.append(0)
    return out
    
y_test_num = convert(y_test)
    
confusion_matrix(y_test_num, predictions.argmax(axis=1))
```

```python
automl.show_models()
```

```python
out = automl.get_models_with_weights()
```

```python
pickle.dump(automl, open('automl.pkl', 'wb'))
pickle.dump(out, open('automl_weights.pkl', 'wb'))
```

# H2O

```python
import h2o
from h2o.automl import H2OAutoML

h2o.init()
```

```python
aml = H2OAutoML(max_models=20, seed=1)
```

```python
df = pd.DataFrame(x_train)
df['Sentiment'] = y_train
```

```python
h2o_data = h2o.import_file('data_files/output_for.csv')

predictors = ['IbmScore', 'GCPScore', 'AzureScore', 'AWSPosScore', 'AWSNegScore', 'AWSNeuScore']
response = "ActualLabel"

train, valid = h2o_data.split_frame(ratios = [.8], seed = 11)
```

```python
train.shape
```

```python
aml.train(x=predictors, y=response, training_frame=h2o_data)
```

```python
lb = aml.leaderboard
lb.head(rows=lb.nrows)
```

```python
predictions = aml.leader.predict(valid)
```

```python
predictions
```

```python
metrics = aml.leader.model_performance(valid)
```

```python
metrics
```

```python
metrics.confusion_matrix()
```

```python
pd = predictions.as_data_frame()
vd = valid.as_data_frame()
```

```python
accuracy_score(vd['ActualLabel'], pd['predict'])
```
