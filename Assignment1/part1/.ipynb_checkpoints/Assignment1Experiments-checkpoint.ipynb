```python
import tensorflow as tf
import keras
from keras.datasets import cifar10, mnist
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.optimizers import RMSprop, Adam, Adagrad, Adadelta, Adamax, Nadam, SGD
import matplotlib.pyplot as plt 
import pickle
import os

# Check if using GPU
from keras import backend as K
K.tensorflow_backend._get_available_gpus()
```

## Pre-processing

```python
(x_train, y_train), (x_test, y_test) = cifar10.load_data()
```

```python
print('x_train shape:', x_train.shape)
print(x_train.shape[0], 'train samples')
print(x_test.shape[0], 'test samples')
print(y_train.shape, 'y shape')
```

```python
x_train = x_train.reshape(50000, 3072)
x_test = x_test.reshape(10000, 3072)
x_train = x_train.astype('float32')
x_test = x_test.astype('float32')
x_train /= 255
x_test /= 255
print(x_train.shape[0], 'train samples')
print(x_test.shape[0], 'test samples')
```

```python
num_classes = 10

# convert class vectors to binary class matrices
y_train = keras.utils.to_categorical(y_train, num_classes)
y_test = keras.utils.to_categorical(y_test, num_classes)
```

## Starting Code
The following is the starter MLP code from https://github.com/fchollet/keras/blob/master/examples/mnist_mlp.py, with the input shape changed to match the CIFAR-10 data.

```python
batch_size = 128
epochs = 20

model = Sequential()
model.add(Dense(512, activation='relu', input_shape=(3072,)))
model.add(Dropout(0.2))
model.add(Dense(512, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(num_classes, activation='softmax'))

model.summary()

model.compile(loss='categorical_crossentropy',
              optimizer=RMSprop(),
              metrics=['accuracy'])

history_start = model.fit(x_train, y_train,
                    batch_size=batch_size,
                    epochs=epochs,
                    verbose=1,
                    validation_data=(x_test, y_test))
score = model.evaluate(x_test, y_test, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

```python
# https://www.learnopencv.com/image-classification-using-feedforward-neural-network-in-keras/

#Plot the Loss Curves
plt.figure(figsize=[8,6])
plt.plot(history_start.history['loss'],'r',linewidth=3.0)
plt.plot(history_start.history['val_loss'],'b',linewidth=3.0)
plt.legend(['Training Loss', 'Validation Loss'],fontsize=14)
plt.xlabel('Epochs ',fontsize=16)
plt.ylabel('Loss',fontsize=16)
plt.title('Loss Curves',fontsize=16)
 
#Plot the Accuracy Curves
plt.figure(figsize=[8,6])
plt.plot(history_start.history['acc'],'r',linewidth=3.0)
plt.plot(history_start.history['val_acc'],'b',linewidth=3.0)
plt.legend(['Training Accuracy', 'Validation Accuracy'],fontsize=14)
plt.xlabel('Epochs ',fontsize=16)
plt.ylabel('Accuracy',fontsize=16)
plt.title('Accuracy Curves',fontsize=16)
```

## Hyperparameter Tuning
### 1. Epochs

```python
experiment_params = [20, 40, 60, 80, 100]
epochs_experiment = []

for epochs in experiment_params:
    print('Training ', str(epochs), ' epochs')
    
    batch_size = 128
    
    model = Sequential()
    model.add(Dense(512, activation='relu', input_shape=(3072,)))
    model.add(Dropout(0.2))
    model.add(Dense(512, activation='relu'))
    model.add(Dropout(0.2))
    model.add(Dense(num_classes, activation='softmax'))

    model.summary()

    model.compile(loss='categorical_crossentropy',
                  optimizer=RMSprop(),
                  metrics=['accuracy'])

    history = model.fit(x_train, y_train,
                        batch_size=batch_size,
                        epochs=epochs,
                        verbose=1,
                        validation_data=(x_test, y_test))
    score = model.evaluate(x_test, y_test, verbose=0)
    print('Test loss:', score[0])
    print('Test accuracy:', score[1])
    print('\n\n')
    
    epochs_experiment.append((str(epochs), model, history))
```

```python
# https://www.learnopencv.com/image-classification-using-feedforward-neural-network-in-keras/

#Plot the Loss Curves
plt.figure(figsize=[8,6])
legend = []
for name, model, history in epochs_experiment:
    
    plt.plot(history.history['loss'],linewidth=1.0)
    plt.plot(history.history['val_loss'],linewidth=1.0, linestyle='dotted')
    legend.append('Training Loss ' + name)
    legend.append('Validation Loss ' + name)

plt.legend(legend,fontsize=14)
plt.xlabel('Epochs ',fontsize=16)
plt.ylabel('Loss',fontsize=16)
plt.title('Loss Curves',fontsize=16)
```

```python
 #Plot the Accuracy Curves
plt.figure(figsize=[8,6])
legend = []
for name, model, history in epochs_experiment:

    plt.plot(history.history['acc'],linewidth=1.0)
    plt.plot(history.history['val_acc'],linewidth=1.0, linestyle='dotted')
    legend.append('Training Accuracy ' + name)
    legend.append('Validation Accuracy ' + name)

plt.legend(legend,fontsize=14)
plt.xlabel('Epochs ',fontsize=16)
plt.ylabel('Accuracy',fontsize=16)
plt.title('Accuracy Curves',fontsize=16)
```

Somewhere between 30 and 40 epochs, the validation scores level off even though the training scores continue to improve, indicating there is overfitting. For now I will use 40 epochs for the remaining experiments.

### 3. Network configuration
### a. Number of neurons
To start, I am going to test doubling and halving the number of neurons, and having the hidden layer with 50% of the number of neurons in the input layer.

```python
experiment_params = [(512, 512), (512, 256), (1024, 1024), (1024, 512), (256, 256)]
neuron_experiment = []

for neuron_a, neuron_b in experiment_params:
    name = '(' + str(neuron_a) + ', ' + str(neuron_b) + ')'
    print('Training ', name, ' neurons')
    
    epochs=40
    batch_size=512
    
    model = Sequential()
    model.add(Dense(neuron_a, activation='relu', input_shape=(3072,)))
    model.add(Dropout(0.2))
    model.add(Dense(neuron_b, activation='relu'))
    model.add(Dropout(0.2))
    model.add(Dense(num_classes, activation='softmax'))

    model.summary()

    model.compile(loss='categorical_crossentropy',
                  optimizer=RMSprop(),
                  metrics=['accuracy'])

    history = model.fit(x_train, y_train,
                        batch_size=batch_size,
                        epochs=epochs,
                        verbose=1,
                        validation_data=(x_test, y_test))
    score = model.evaluate(x_test, y_test, verbose=0)
    print('Test loss:', score[0])
    print('Test accuracy:', score[1])
    print('\n\n')
    
    neuron_experiment.append((name, model, history))
```

### 4. Learning rate

```python
experiment_params = [.0001, .0005, .001, .005, .01, .05]
lr_experiment = []

for lr in experiment_params:
    print('Training ', str(lr), ' learning rate')
    epochs=40
    batch_size=512

    model = Sequential()
    model.add(Dense(1024, activation='relu', input_shape=(3072,)))
    model.add(Dropout(0.2))
    model.add(Dense(512, activation='relu'))
    model.add(Dropout(0.2))
    model.add(Dense(256, activation='relu'))
    model.add(Dropout(0.2))
    model.add(Dense(num_classes, activation='softmax'))

    model.summary()

    model.compile(loss='categorical_crossentropy',
                  optimizer=RMSprop(lr=lr),
                  metrics=['accuracy'])

    history = model.fit(x_train, y_train,
                        batch_size=batch_size,
                        epochs=epochs,
                        verbose=1,
                        validation_data=(x_test, y_test))
    score = model.evaluate(x_test, y_test, verbose=0)
    print('Test loss:', score[0])
    print('Test accuracy:', score[1])
    print('\n\n')

    lr_experiment.append((str(lr), model, history))
```

```python
#Plot the Loss Curves
plt.figure(figsize=[8,6])
legend = []
for name, model, history in lr_experiment:
    
    plt.plot(history.history['loss'],linewidth=1.0)
    plt.plot(history.history['val_loss'],linewidth=1.0, linestyle='dotted')
    legend.append('Training Loss ' + name)
    legend.append('Validation Loss ' + name)

plt.legend(legend,fontsize=14)
plt.xlabel('Epochs ',fontsize=16)
plt.ylabel('Loss',fontsize=16)
plt.title('Loss Curves',fontsize=16)
```

```python
 #Plot the Accuracy Curves
plt.figure(figsize=[8,6])
legend = []
for name, model, history in lr_experiment:

    plt.plot(history.history['acc'],linewidth=1.0)
    plt.plot(history.history['val_acc'],linewidth=1.0, linestyle='dotted')
    legend.append('Training Accuracy ' + name)
    legend.append('Validation Accuracy ' + name)

plt.legend(legend,fontsize=14)
plt.xlabel('Epochs ',fontsize=16)
plt.ylabel('Accuracy',fontsize=16)
plt.title('Accuracy Curves',fontsize=16)
```

### 5. Activation function
In exploring the activation functions, I will first look at their affects on the output layer.

```python
experiment_params = ['softmax','elu','selu','softplus', 'softsign','relu', 
                     'tanh', 'sigmoid', 'hard_sigmoid', 'exponential', 'linear']
activation_experiment = []

for activation in experiment_params:
    print('Training ', activation, ' activation function')
    epochs=40
    batch_size=512

    model = Sequential()
    model.add(Dense(1024, activation='relu', input_shape=(3072,)))
    model.add(Dropout(0.2))
    model.add(Dense(512, activation='relu'))
    model.add(Dropout(0.2))
    model.add(Dense(256, activation='relu'))
    model.add(Dropout(0.2))
    model.add(Dense(num_classes, activation=activation))

    model.summary()

    model.compile(loss='categorical_crossentropy',
                  optimizer=RMSprop(),
                  metrics=['accuracy'])

    history = model.fit(x_train, y_train,
                        batch_size=batch_size,
                        epochs=epochs,
                        verbose=1,
                        validation_data=(x_test, y_test))
    score = model.evaluate(x_test, y_test, verbose=0)
    print('Test loss:', score[0])
    print('Test accuracy:', score[1])
    print('\n\n')

    activation_experiment.append((activation, model, history))
```

### 6. Dropout
First, I will take a look at what happens if there is no dropout. THen I will look at what happens when I increment the dropout at a rate of .05.

```python
epochs=40
batch_size=512

model = Sequential()
model.add(Dense(1024, activation='relu', input_shape=(3072,)))
#model.add(Dropout(.2))
model.add(Dense(512, activation='relu'))
#model.add(Dropout(.2))
model.add(Dense(256, activation='relu'))
#model.add(Dropout(.2))
model.add(Dense(num_classes, activation='softmax'))

model.summary()

model.compile(loss='categorical_crossentropy',
              optimizer=RMSprop(),
              metrics=['accuracy'])

history1 = model.fit(x_train, y_train,
                    batch_size=batch_size,
                    epochs=epochs,
                    verbose=1,
                    validation_data=(x_test, y_test))
score = model.evaluate(x_test, y_test, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

```python
#Plot the Loss Curves
plt.figure(figsize=[8,6])
plt.plot(history1.history['loss'],'r',linewidth=3.0)
plt.plot(history1.history['val_loss'],'b',linewidth=3.0)
plt.legend(['Training Loss', 'Validation Loss'],fontsize=14)
plt.xlabel('Epochs ',fontsize=16)
plt.ylabel('Loss',fontsize=16)
plt.title('Loss Curves',fontsize=16)
 
#Plot the Accuracy Curves
plt.figure(figsize=[8,6])
plt.plot(history1.history['acc'],'r',linewidth=3.0)
plt.plot(history1.history['val_acc'],'b',linewidth=3.0)
plt.legend(['Training Accuracy', 'Validation Accuracy'],fontsize=14)
plt.xlabel('Epochs ',fontsize=16)
plt.ylabel('Accuracy',fontsize=16)
plt.title('Accuracy Curves',fontsize=16)
```

```python
experiment_params = [.05, .1, .15, .2, .25, .3]
dropout_experiment = []

for dropout in experiment_params:
    print('Training ', dropout, ' dropout rate')
    epochs=40
    batch_size=512

    model = Sequential()
    model.add(Dense(1024, activation='relu', input_shape=(3072,)))
    model.add(Dropout(dropout))
    model.add(Dense(512, activation='relu'))
    model.add(Dropout(dropout))
    model.add(Dense(256, activation='relu'))
    model.add(Dropout(dropout))
    model.add(Dense(num_classes, activation='softmax'))

    model.summary()

    model.compile(loss='categorical_crossentropy',
                  optimizer=RMSprop(),
                  metrics=['accuracy'])

    history = model.fit(x_train, y_train,
                        batch_size=batch_size,
                        epochs=epochs,
                        verbose=1,
                        validation_data=(x_test, y_test))
    score = model.evaluate(x_test, y_test, verbose=0)
    print('Test loss:', score[0])
    print('Test accuracy:', score[1])
    print('\n\n')

    dropout_experiment.append((str(dropout), model, history))
```

```python
#Plot the Loss Curves
plt.figure(figsize=[8,6])
legend = []
for name, model, history in dropout_experiment:
    
    plt.plot(history.history['loss'],linewidth=1.0)
    plt.plot(history.history['val_loss'],linewidth=1.0, linestyle='dotted')
    legend.append('Training Loss ' + name)
    legend.append('Validation Loss ' + name)

plt.legend(legend,fontsize=10)
plt.xlabel('Epochs ',fontsize=16)
plt.ylabel('Loss',fontsize=16)
plt.title('Loss Curves',fontsize=16)
```

```python
 #Plot the Accuracy Curves
plt.figure(figsize=[8,6])
legend = []
for name, model, history in dropout_experiment:

    plt.plot(history.history['acc'],linewidth=1.0)
    plt.plot(history.history['val_acc'],linewidth=1.0, linestyle='dotted')
    legend.append('Training Accuracy ' + name)
    legend.append('Validation Accuracy ' + name)

plt.legend(legend,fontsize=10)
plt.xlabel('Epochs ',fontsize=16)
plt.ylabel('Accuracy',fontsize=16)
plt.title('Accuracy Curves',fontsize=16)
```

The smaller dropout rates appear to be overfitting, though not to the same degree as when there was no dropout. Let's look at the ones where the rate is >= 0.2

```python
#Plot the Loss Curves
plt.figure(figsize=[8,6])
legend = []
for name, model, history in dropout_experiment:
    if float(name) >= 0.2:
        plt.plot(history.history['loss'],linewidth=1.0)
        plt.plot(history.history['val_loss'],linewidth=1.0, linestyle='dotted')
        legend.append('Training Loss ' + name)
        legend.append('Validation Loss ' + name)

plt.legend(legend,fontsize=10)
plt.xlabel('Epochs ',fontsize=16)
plt.ylabel('Loss',fontsize=16)
plt.title('Loss Curves',fontsize=16)

#Plot the Accuracy Curves
plt.figure(figsize=[8,6])
legend = []
for name, model, history in dropout_experiment:
    if float(name) >= 0.2:
        plt.plot(history.history['acc'],linewidth=1.0)
        plt.plot(history.history['val_acc'],linewidth=1.0, linestyle='dotted')
        legend.append('Training Accuracy ' + name)
        legend.append('Validation Accuracy ' + name)

plt.legend(legend,fontsize=10)
plt.xlabel('Epochs ',fontsize=16)
plt.ylabel('Accuracy',fontsize=16)
plt.title('Accuracy Curves',fontsize=16)
```

The rate of 0.2 gives the highest accuracy and lowest loss, so I will continue to use that.

### Revisiting Epochs
As the first step in this set of experiments, the number of epochs was selected for a very different architecture. Given that the model is not overfitting at 40 epochs, it is worth revisiting the question of how many epochs to use.

```python
experiment_params = [40 + 10*i for i in range(5)]
epoch2_experiment = []

for epochs in experiment_params:
    print('Training ', epochs, ' epochs')

    batch_size=512

    model = Sequential()
    model.add(Dense(1024, activation='relu', input_shape=(3072,)))
    model.add(Dropout(0.2))
    model.add(Dense(512, activation='relu'))
    model.add(Dropout(0.2))
    model.add(Dense(256, activation='relu'))
    model.add(Dropout(0.2))
    model.add(Dense(num_classes, activation='softmax'))

    model.summary()

    model.compile(loss='categorical_crossentropy',
                  optimizer=RMSprop(),
                  metrics=['accuracy'])

    history = model.fit(x_train, y_train,
                        batch_size=batch_size,
                        epochs=epochs,
                        verbose=1,
                        validation_data=(x_test, y_test))
    score = model.evaluate(x_test, y_test, verbose=0)
    print('Test loss:', score[0])
    print('Test accuracy:', score[1])
    print('\n\n')

    epoch2_experiment.append((str(dropout), model, history))
```
